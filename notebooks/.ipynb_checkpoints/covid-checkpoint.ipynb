# Plot cases by date for Afghanistan

```python
''' Plot the individual year time series'''

''' change y_var to view the desired time series '''



filtered = _grp_book_df[_grp_book_df['countriesAndTerritories'] == 'Afghanistan']

y_var = 'dateRep'   # set the dependent variable (e.g. RoomNights, AgentCost)

x_var = 'cases'

ax = filtered.plot(y_var, x_var)

ax.set_title("Daily Total cases", fontsize=20)
ax.set_xlabel("Date",fontsize=16)
ax.set_ylabel("Cases", fontsize=16)
ax.legend(fontsize=16)
```
